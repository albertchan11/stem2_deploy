## Stage 1
A little bot is lost in the maze. It is surrounded by walls with only one way to go. You will need to help it find its way out.

Below is the picture of the maze. The gray square is the poor bot and black square is the exit. (You will earn 40 points for helping the bot reach the exit.)

### Functions you can use

- `bot.move(direction)`: Ask the bot to move one step to the direction if there is no wall
- `bot.can_move(direction)`: Tell if there is a wall in the direction by returning a Boolean value. `True` if there is no wall; `False` if there is a wall

The value for direction can be `Direction.UP`, `Direction.LEFT`, `Direction.DOWN`, `Direction.RIGHT`

### Restrictions
- In the while loop, for each iteration, the bot is allowed to move only one step
- Find the exit within 15 steps

Please run the below block to generate the maze

```python
maze1 = generate_2d_maze(maze1_str)
draw_maze(maze1)
```

```python
bot = Bot(Direction.UP, maze1)
counter = 0

while counter < 15:
    # -------ADD YOUR CODE BELOW-------
    # Your Implementation
    # -------ADD YOU CODE ABOVE-------

    if not only_one_step(bot):
        print("You cannot ask the bot to move two steps in one round. It's gonna be exhausting 3_3)")
        break

    draw_maze(maze1, bot)
    counter += 1

    if finish(maze1, bot):
        print("Thank you! You are such a kind person to help the bot! Here is your 40 points.")
        break

if not finish(maze1, bot):
    print("Sorry you failed to help the bot...")
```
